```python
df_sales=pd.read_csv(r'C:\Users\udeva\OneDrive\Documents\K.ineuron\sales_data_final.csv')
```

```python
def profit_flag(a):
    if a>0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(-45)
```

```python
profit_flag(45)
```

```python
df_sales['profit_flag']=df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
df_sales['length']=df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def quantity_flag(a):
    if a < 10:
        return 'low'
    elif a>10:
        return 'medium'
    else:
        return 'high'
```

```python
df_sales['quantity_flag']=df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
a=df_sales
```

```python
df_sales['square_quantity']=df_sales['quantity'].apply(lambda a:a*a)
```
